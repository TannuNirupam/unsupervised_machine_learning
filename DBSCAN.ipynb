```python
import pandas as pd 

df= pd.read_csv("blobs.csv")
df.head()
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

sns.scatterplot(df, x="0",y="1")
plt.show()
```

```python
from sklearn.cluster import DBSCAN
ds = DBSCAN (eps=1, min_samples=5)
ds.fit(df[['0','1']])
list(ds.labels_).count(-1)
```

```python
df['cluster']=ds.labels_
df.head()
```

```python
df.cluster.value_counts()
```

```python
sns.scatterplot(df,x='0' , y='1' , palette="tab10" , s= 100, hue="cluster")
plt.show()
```
